# Climate Analysis by Tim Lucas

The primary purpose of this notebook is to analyze percipitation data as well as the stations that are reporting it. This data is being analyzed for use in a python flask application.

```python
#Import dependencies
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
from datetime import timedelta
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# Reflect an existing database into a new model.
Base = automap_base()

# Reflect the tables.
Base.prepare(engine, reflect=True)
```

```python
# View of all of the classes that automap found.
Base.classes.keys()
```

```python
# Saving references to each table.
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Creating our session (link) from Python to the DB.
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Finding the last measuremetn record and the date 12 months prior to that.
last_date = session.query(Measurement).order_by(Measurement.date.desc()).first()
last_12 = (dt.datetime.strptime(last_date.date, '%Y-%m-%d') - timedelta(days=365)).strftime("%Y-%m-%d")

# Query for finding the precipitation data for last 12 months from last measurement date.
prec_data = session.query(Measurement).filter(Measurement.date > last_12).all()
date = []
prcp = []

for prec in prec_data:
    date.append(prec.date)
    prcp.append(prec.prcp)

# Query results put into a Pandas DataFrame and the index set to the date column.
measurement_df = pd.DataFrame({
    'date': date,
    'prcp': prcp
})
measurement_df = measurement_df.set_index('date')

# Sorting the dataframe by date.
measurement_df = measurement_df.sort_values(['date'])

# Pandas Plotting with Matplotlib for precipitation data.
plt.figure(figsize=(10,5))
plt.tight_layout()
plt.plot(measurement_df['prcp'])
plt.title("12 Months Precipitation by Date")
plt.xlabel("Date")
plt.ylabel("Inches")
plt.show()
```

```python
# Summary statistics for percipitation data usingn pandas.
measurement_df.describe()
```

```python
# The total number of stations available in the dataset.
station_count = session.query(func.count(Station.name)).first()
station_count
```

```python
# List of the stations and the measurement counts in descending order.
station_rows = session.query(Station.station, Station.name, func.count(Station.name)).filter(Station.station == Measurement.station).group_by(Station.name).order_by(func.count(Station.name).desc()).all()
station_rows
```

```python
# The lowest temperature recorded, highest temperature recorded, average temperature of the most active station, USC00519281.
active_station = session.query(Measurement.station, func.min(Measurement.prcp), func.max(Measurement.prcp), func.avg(Measurement.prcp)).filter(Measurement.station == 'USC00519281').group_by(Measurement.station).all()
active_station
```

```python
# USC00519281 is the station with the highest number of obsrvations.
# This is the last 12 months of temperature observation data for this station and a plot of the results as a histogram.
station_data = session.query(Measurement).filter(Measurement.station == 'USC00519281').filter(Measurement.date > last_12).all()
tobs = []

for row in station_data:
    tobs.append(row.tobs)

# Save the query results as a Pandas DataFrame and set the index to the date column
station_df = pd.DataFrame({
    'tobs': tobs
})
```

```python
# Histogram plotted with pandas.
station_df.hist(bins=12)
plt.title("12 Months TOBS for USC00519281")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()
```
